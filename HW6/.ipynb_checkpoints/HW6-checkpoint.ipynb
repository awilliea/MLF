```python
import sklearn.svm as svm
import numpy as np 
import random
```

```python
train_x = []
train_y = []
with open('train.txt') as f:
    for d in f.readlines():
        d = d.strip().split(' ')
        train_x.append([float(d[3]),float(d[5])])
        train_y.append(int(float(d[0])))
train_x = np.array(train_x)
train_y = np.array(train_y)
```

```python
test_x = []
test_y = []
with open('test.txt') as f:
    for d in f.readlines():
        d = d.strip().split(' ')
        test_x.append([float(d[3]),float(d[5])])
        test_y.append(int(float(d[0])))
test_x = np.array(test_x)
test_y = np.array(test_y)
```

```python
def adjust(train_y,test_y,target):
    return (train_y == target).astype(int),  (test_y == target).astype(int)   
```

```python
train_y_2, test_y_2 = adjust(train_y,test_y,2)
w_list = []

for c in [i**10 for i in [-5,-3,-1,1,3]]:
    clf = svm.SVC(kernel = 'linear',shrinking=False,C=c)
    clf.fit(train_x,train_y_2)
    w = clf.coef_[0]
    w_list.append(np.linalg.norm(w))
```
